```python
def analisar_colunas_numericas(df):
  """
  Analisa colunas numéricas (float) em um DataFrame.

  Argumentos:
    df: DataFrame do Pandas.

  Retorno:
    DataFrame com colunas numéricas preenchidas com média ou mediana.
  """

  # Identificando colunas numéricas
  colunas_numericas = list(df.select_dtypes(include=['float']).columns)

  # Percorrer cada coluna numérica
  for coluna in colunas_numericas:
    # Verificar se há dados faltantes
    if df[coluna].isnull().any():
      # Calcular média e mediana
      media = df[coluna].mean()
      mediana = df[coluna].median()

      # Decidir qual valor usar para preencher os dados faltantes
      valor_preenchimento = media if abs(media - mediana) < abs(df[coluna].kurtosis()) else mediana

      # Preencher os dados faltantes
      df[coluna].fillna(np.nan, inplace=True)

  return df

# Exemplo de uso
df = analisar_colunas_numericas(df.copy())
```

```python
def tratar_outliers(df, colunas_numericas):
    """
    Trata outliers em colunas numéricas de um DataFrame.

    Argumentos:
      df: DataFrame do Pandas.
      colunas_numericas: Lista de colunas numéricas.

    Retorno:
      DataFrame com outliers removidos ou substituídos pela média.
    """

    # Percorrer cada coluna numérica
    for coluna in colunas_numericas:
        # Calcular média e desvio padrão
        media = df[coluna].mean()
        desvio_padrao = df[coluna].std()
        
        # Calcular limites do intervalo de confiança
        limite_inferior = media - 2.7 * desvio_padrao
        limite_superior = media + 2.7 * desvio_padrao
        
        # Substituir outliers por NaN e remover números negativos
        df[coluna] = np.where((df[coluna] < limite_inferior) | (df[coluna] > limite_superior) | (df[coluna] < 0), np.nan, df[coluna])
    return df

# Exemplo de uso
colunas_numericas = list(df.select_dtypes(include=['float64']).columns)
df = tratar_outliers(df.copy(), colunas_numericas)
```

```python

# Calcular os quartis do subtotal
q1_subtotal = df['subtotal'].quantile(0.25)
q2_subtotal = df['subtotal'].quantile(0.5)
q3_subtotal = df['subtotal'].quantile(0.75)

# Definir função para atribuir cores com base nos quartis
def assign_color(subtotal):
    if subtotal < q1_subtotal:
        return 'Q1'
    elif subtotal < q2_subtotal:
        return 'Q2'
    elif subtotal < q3_subtotal:
        return 'Q3'
    else:
        return 'Q4'

# Adicionar coluna de cor com base nos quartis
df['subtotal_quartile'] = df['subtotal'].apply(assign_color)

# Plotar o gráfico de dispersão com cores por quartil de subtotal
plt.figure(figsize=(10, 6))
sns.scatterplot(x='subtotal', y='estimated_store_to_consumer_driving_duration', hue='subtotal_quartile', data=df, palette='Set2')

# Ajustar a escala dos eixos x e y
# Substitua os valores de xmin, xmax, ymin, ymax pelos limites desejados
xmin, xmax = df['subtotal'].min(), df['subtotal'].max()
ymin, ymax = df['estimated_store_to_consumer_driving_duration'].min(), df['estimated_store_to_consumer_driving_duration'].max()
plt.xlim(xmin, xmax*1.2)
plt.ylim(ymin, ymax*1.5)

plt.xlabel('Subtotal')
plt.ylabel('Tempo de Entrega Estimado (s)')
plt.title('Subtotal vs. Tempo de Entrega Estimado')
plt.legend(title='Quartil de Subtotal', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Calculate quartiles for 'awaited_time'
quartiles = df['awaited_time'].quantile([0.25, 0.5, 0.75])
q1, q2, q3 = quartiles[0.25], quartiles[0.5], quartiles[0.75]

# Assign colors based on quartiles
colors = np.where(df['awaited_time'] <= q1, 'blue', 
            np.where(df['awaited_time'] <= q2, 'green',
            np.where(df['awaited_time'] <= q3, 'orange', 'red')))

# Create the scatter plot
plt.figure(figsize=(10, 6))
scatter = plt.scatter( df['total_busy_dashers'],df['total_outstanding_orders'], c=colors, alpha=0.5)
plt.colorbar(scatter, label='Awaited Time')
plt.ylabel('Total Outstanding Orders')
plt.xlabel('Total Busy Dashers')
plt.title('Scatter Plot of Pending Deliveries, Busy Dashers, and Awaited Time')

plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import matplotlib.pyplot as plt

# Supondo que df é o seu DataFrame
# Removendo linhas com valores NaN
df.dropna(inplace=True)

# Definindo X e y após remover linhas com NaN
X = df[['total_items', 'subtotal', 'total_onshift_dashers', 'total_busy_dashers', 'total_outstanding_orders', 'estimated_store_to_consumer_driving_duration']]
y = df['awaited_time']

# Dividindo os dados
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Criando o modelo de regressão linear
regressor = LinearRegression()

# Treinando o modelo
regressor.fit(X_train, y_train)

# Fazendo previsões
y_pred = regressor.predict(X_test)

# Calculando os erros
mae = metrics.mean_absolute_error(y_test, y_pred)
mse = metrics.mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print('Mean Absolute Error:', mae)
print('Mean Squared Error:', mse)
print('Root Mean Squared Error:', rmse)

# Plotando o gráfico
plt.scatter(y_test, y_pred)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=4)
plt.xlabel('Actual Delivery Time')
plt.ylabel('Predicted Delivery Time')
plt.title('Actual vs Predicted Delivery Time')
plt.show()
```
